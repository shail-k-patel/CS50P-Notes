# 2 Graph in 1

```python
# Data for the first line
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])

# Data for the second line
x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])

# Plot the first line
plt.plot(x1, y1, label="Line 1: y1 vs x1", color="blue", marker="o")

# Plot the second line
plt.plot(x2, y2, label="Line 2: y2 vs x2", color="red", linestyle="--", marker="s")

# Add labels and legend
plt.title("Two Lines in One Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Show the legend
plt.grid()    # Add a grid

# Show the plot
plt.show()
```

# Nested Graph

```python
# Main plot data
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]

# sub plot data
x2 = [1, 2, 3, 4]
y2 = [16, 9, 4, 1]

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # Main plot
axes2 = fig.add_axes([0.2,0.6,0.2,0.3]) # Sub plot

axes1.plot(x1,y1,'r')
axes2.plot(x2,y2,'g')

plt.show()
```

# ColorMap
A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

# here color are independent ofx and y they are just according to colors above and order in which they appear

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar() # shows color bar on side

plt.show()
```

# Question
x and y are randomly generated values between 0 and 50. <br>
The color of each point is determined by the sum of x + y, which ranges between 0 and 100. <br>
This sum is mapped to the colormap (viridis) to decide the color. <br>

```python
# Generate random values for x and y, both between 0 and 50
x = np.random.uniform(0, 50, 100)  # 100 random x-values between 0 and 50
y = np.random.uniform(0, 50, 100)  # 100 random y-values between 0 and 50

# Calculate the color of each point based on the sum of x + y
# Sum ranges between 0 and 100 since x and y are each between 0 and 50
colors = x + y  

# Plot scatter plot
plt.scatter(x, y, c=colors, cmap='Purples')  # Use 'viridis' colormap to map color
plt.colorbar(label="x + y")  # Add colorbar to show the mapping of values to colors

# Add labels and title for clarity
plt.xlabel("x values (0 - 50)")
plt.ylabel("y values (0 - 50)")
plt.title("Scatter Plot: Color Determined by x + y (0 - 100)")

plt.show()
```

# fill_between

```python
x = np.linspace(0, 10, 100)  # Create x values
y = np.sin(x)  # Compute y values

plt.plot(x, y, label="sin(x)")
plt.fill_between(x, y, color="blue", alpha=0.3)  # Fill under the curve
plt.title("Filling Under a Curve")
plt.legend()
plt.show()
```

# 3D

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.random.rand(50) * 10  # Random x values
y = np.random.rand(50) * 10  # Random y values
z = np.random.rand(50) * 10  # Random z values

ax = plt.axes(projection='3d')

# Scatter plot
scatter = ax.scatter(x, y, z)


# Set axis labels
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

# Set title
ax.set_title("3D Scatter Plot")

plt.show()
```

```python
# Generate grid data
x = np.linspace(-5, 5, 50)  # x-axis values
y = np.linspace(-5, 5, 50)  # y-axis values
X, Y = np.meshgrid(x, y)  # Create a 2D grid
Z = np.sin(np.sqrt(X**2 + Y**2))  # Define a function for z-values

# Create a 3D surface plot
ax = plt.axes(projection='3d')  # Directly create a 3D axes

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Set axis labels and title
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("3D Surface Plot")

plt.show()
```
